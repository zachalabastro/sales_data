## Question 3

```python
import pandas as pd
import matplotlib.pyplot as plt

df_1 = pd.read_csv('fct_invoice.csv')

df_1
```

```python
df_1.sort_values('invoice_date')
```

```python
frequency = df_1.value_counts('invoice_date')

frequency
```

```python
for_graph = frequency.sort_index()

for_graph
```

```python
plt.plot(for_graph.index, for_graph.values)
plt.xlabel('Invoice Date')
plt.ylabel('Frequency')
plt.title('Frequency of Dates')
plt.show()
```

```python
def get_season(date):
    month_day = date.strftime('%m-%d')
    seasons = {'Spring': [('03-01', '05-31')],
               'Summer': [('06-01', '08-31')],
               'Autumn': [('09-01', '11-30')],
               'Winter': [('12-01', '12-31'), ('01-01', '02-28')]}
    for season, date_ranges in seasons.items():
        for date_range in date_ranges:
            start, end = date_range
            if start <= month_day <= end:
                return season
        
df_1['season'] = df_1['invoice_date'].apply(get_season)

df_1
```

```python
df_1.sort_values('invoice_date')
```

```python
season_graph = df_1.value_counts('season')

season_graph
```

```python
plt.plot(season_graph.index, season_graph.values)
plt.xlabel('Season')
plt.ylabel('Frequency')
plt.title('Frequency of Seasons')
plt.show()
```

```python
#During the winter season (whatever the year is), the number of people buying is at its peak. This may be because of the numerous holidays and celebrations such as Christmas and New Year present during this season.
```

```python
import pandas as pd

df_1 = pd.read_csv('fct_invoice.csv')
df_2 = pd.read_json('dim_customer.json')

complete_df = pd.merge(
    df_1,
    df_2,
    how='inner',
    left_on='customer_id',
    right_on='id'
)

complete_df
```

```python
def decade_age(age):
    return f'{(age // 10) * 10}-{(age // 10) * 10 + 9}'

complete_df['decade_age'] = complete_df['age'].apply(decade_age)

complete_df
```

```python
complete_df.sort_values('decade_age')
```

```python
pivot_table = pd.pivot_table(complete_df, values='price', index='category', columns='decade_age', aggfunc='sum')

pivot_table
```
